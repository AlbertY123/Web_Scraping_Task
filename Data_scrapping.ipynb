## Step 2: Import Libraries
Import the required Python libraries.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Step 3: Send a Request to the Website
Send an HTTP request to the website and retrieve the HTML content.

```python
# Define the URL of the website
url = "http://books.toscrape.com/"

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print("Successfully connected to the website!")
else:
    print("Failed to connect to the website.")
```

## Step 4: Parse the HTML Content
Use BeautifulSoup to parse the HTML content and extract data.

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")

# Print the title of the webpage to verify parsing
print("Page Title:", soup.title.text)
```

## Step 5: Inspect the Website Structure
Before scraping, inspect the website structure using your browser’s Developer Tools (right-click → Inspect). Identify the HTML elements containing the book data (e.g., titles, prices, ratings).

For example:

- Book titles are inside \<h3\> tags.

- Prices are inside \<p class="price_color"\> tags.

- Ratings are inside \<p class="star-rating"\> tags.

## Step 6: Extract Book Data
Extract the book titles, prices, and ratings from the webpage.

```python
# Find all book containers (each book is inside an <article> tag)
books = soup.find_all("article", class_="product_pod")

# Initialize lists to store the data
titles = []
prices = []
ratings = []

# Loop through each book and extract the data
for book in books:
    # Extract the title
    title = book.h3.a["title"]
    titles.append(title)
    
    # Extract the price
    price = book.find("p", class_="price_color").text
    prices.append(price)
    
    # Extract the rating
    rating = book.p["class"][1]  # The second class name indicates the rating (e.g., "Three")
    ratings.append(rating)

# Print the first few entries to verify
print("Titles:", titles[:5])
print("Prices:", prices[:5])
print("Ratings:", ratings[:5])
```

## Step 7: Save the Data to a DataFrame
Store the extracted data in a Pandas DataFrame for further analysis.

```python
# Create a DataFrame
data = {
    "Title": titles,
    "Price": prices,
    "Rating": ratings
}

df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print(df.head())
```

## Step 8: Save the Data to a CSV File
Save the scraped data to a CSV file for later use.

```python
# Save the DataFrame to a CSV file
df.to_csv("books_data.csv", index=False)

print("Data saved to books_data.csv!")
```
